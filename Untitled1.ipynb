```python
import pyaudio
import wave
import speech_recognition as sr

def audio_save():

    FORMAT = pyaudio.paInt16  # 16ë¹„íŠ¸ í˜•ì‹ìœ¼ë¡œ ì„¤ì •
    CHANNELS = 1
    RATE = 44100
    CHUNK = 1024
    RECORD_SECONDS = 10
    OUTPUT_FILENAME = "recorded_audio.wav"  # ì €ì¥í•  íŒŒì¼ ì´ë¦„


    audio = pyaudio.PyAudio()

    stream = audio.open(format=FORMAT, channels=CHANNELS,
                        rate=RATE, input=True,
                        frames_per_buffer=CHUNK)

    print("ë…¹ìŒ ì‹œì‘...")

    frames = []
    for _ in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(data)

    print("ë…¹ìŒ ì™„ë£Œ.")


    stream.stop_stream()
    stream.close()
    audio.terminate()

    # ìŒì„± íŒŒì¼ë¡œ ì €ì¥
    with wave.open(OUTPUT_FILENAME, 'wb') as wf:
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(audio.get_sample_size(FORMAT))
        wf.setframerate(RATE)
        wf.writeframes(b''.join(frames))

    return print("íŒŒì¼ ì €ì¥ ì™„ë£Œ:", OUTPUT_FILENAME)

def tts_module():

    r = sr.Recognizer()
    kr_audio = sr.AudioFile('recorded_audio.wav')

    with kr_audio as source:
        audio = r.record(source)

    a=r.recognize_google(audio, language='ko-KR')
    print(r.recognize_google(audio, language='ko-KR'))



    return a

if __name__ == '__main__':
    audio_save()
    sentiment_data=tts_module()
    print("a",sentiment_data)
```
